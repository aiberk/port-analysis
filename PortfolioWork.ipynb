```python
import yfinance as yf

ticker = yf.Ticker("GC=F")
df = ticker.history(start="2018-01-01", end="2023-12-31")

# Calculate the daily returns
df['returns'] = df['Close'].pct_change()

# Calculate the moving average
df['sma'] = df['Close'].rolling(window=20).mean()

# Calculate the Bollinger bands
df['upper_band'] = df['sma'] + 2 * df['returns'].rolling(window=20).std()
df['lower_band'] = df['sma'] - 2 * df['returns'].rolling(window=20).std()

# Plot the data
df[['Close', 'sma', 'upper_band', 'lower_band']].plot(title='Chocolate Futures')

# Print the last 5 rows of data
print(df.tail())
```
